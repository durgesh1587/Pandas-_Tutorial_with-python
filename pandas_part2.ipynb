```python
data1=pd.DataFrame({'key' :[1,2,3,4,5],
       'key1':[54,57,24,85,42],
       'key2':[7,8,9,4,5]
      })
```

```python
data1
```

```python
data2=pd.DataFrame({'key' :[1,2,3,4,5],
       'key3':[5,5,2,8,4],
       'key4':[7,8,9,4,5]
      })
```

```python
data2
```

```python
#* inner: use intersection of keys from both frames, similar to a SQL inner
# join; preserve the order of the left keys.
pd.merge(data1,data2)
```

```python
#* left: use only keys from left frame, similar to a SQL left outer join;
#preserve key order.
pd.merge(data1,data2,how='left')
```

```python
#* right: use only keys from right frame, similar to a SQL right outer join;
# preserve key order.
pd.merge(data1,data2,how='right')
```

```python
#outer: use union of keys from both frames, similar to a SQL full outer
#join; sort keys lexicographically
pd.merge(data1,data2,how='outer')
```

```python
# cross:creates the cartesian product from both frames, preserves the order
#of the left keys.
pd.merge(data1,data2,how='cross')
```

```python
# this function basicly using col wise marge and fing data "on"function
pd.merge(data1,data2,how='outer',on ='key')
```
